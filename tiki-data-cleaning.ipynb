Thoạt nhìn từ dữ liệu, ta có thể thấy một số vấn đề xuất hiện ngay trên bề nồi.
- Các cột <b> 'authors', 'quantity', 'pages' và 'manufacturer'</b> hiện đang có nhiều giá trị null.
- Cột <b>'quantity'</b> bị sai kiểu dữ liệu (là float trong số khi nên là int), <b>'pages'</b> bị sai kiểu dữ liệu (là object trong khi nên là int)
- Các dòng dữ liệu bị lặp.
- Các cột chưa về đúng dạng tên chính xác.

Ta sẽ đổi kiểu dữ liệu, đổi tên cột và làm việc với vấn đề giá trị null của các cột trước.

### 1.1. Cột 'quantity'

```python
tiki_work["quantity"]
```

```python
#Đến dòng này cần skip qua dòng tiếp theo chạy tiếp, do dòng này cố tình chạy lỗi để mô tả ý tưởng
tiki_work["quantity"].astype('int64')
```

Cột 'quantity' hiện chưa thể chuyển đổi do vẫn còn giá trị null. Tuy vậy, ta có thể hiểu null ở đây là chưa bán được sách => số lượng = 0.

Do vậy, trong cột này, bất cứ giá trị nào bị null cũng sẽ được thay thế bởi 0.

```python
tiki_work["quantity"].isna()
```

```python
tiki_work["quantity"] = tiki_work["quantity"].mask(tiki_work["quantity"].isna() == True, 0)
tiki_work["quantity"]
```

```python
test = tiki_book["quantity"].copy()
```

```python
#Kiểm tra những giá trị nào đang là 0 bằng biến test
test = test.value_counts().reset_index()
test = pd.DataFrame(test)
```

```python
test
```

Có 45 giá trị null đã được chuyển thành 0, giờ ta sẽ kiểm tra lại cột một lần nữa để đảm bảo không còn giá trị null.

```python
tiki_work.info()
```

Như ta thấy, cột 'quantity' đã không còn giá trị null, giờ ta có thể đổi kiểu dữ liệu về dạng int.

```python
tiki_work["quantity"] = tiki_work["quantity"].astype('int64')
```

```python
#Kiểm tra kiểu dữ liệu 
tiki_work.info()
```

```python
tiki_work['quantity']
```

### 1.2. Cột 'authors'

```python
tiki_work['authors'].unique()
```

Từ cột này tới các cột về sau sẽ không giải thích code tiếp do quy trình đã tương đối rõ ràng.

```python
tiki_work['authors'].isna().sum()
```

```python
tiki_work['authors'] = tiki_work['authors'].mask(tiki_work['authors'].isna(), 'Unknown')
tiki_work['authors'].isna().sum()
```

```python
tiki_work['authors']
```

```python
tiki_work.info()
```

### 1.3. Cột 'pages'
Riêng cột này ta cần cách xử lý khác với cột <b>'quantity'</b>, do một phép tính sai có thể ảnh hưởng tới phân tích chung. Ta sẽ fill giá trị null với giá trị mean (trung bình) của cột.

```python
tiki_work['pages']
```

```python
tiki_work['pages'] = tiki_work['pages'].replace('Cuốn', 'NaN')
tiki_work['pages']
```

```python
tiki_work['pages'] = tiki_work['pages'].astype('float64')
tiki_work['pages']
```

```python
fill_mean = tiki_work['pages'].mean().round(0)
fill_mean
```

```python
tiki_work['pages'] = tiki_work['pages'].fillna(value=fill_mean)
```

```python
tiki_work['pages'] = tiki_work['pages'].astype('int64')
tiki_work['pages']
```

### 1.4. Cột 'manufacturer'

```python
tiki_work['manufacturer'].isna().sum()
```

```python
tiki_work['manufacturer'] = tiki_work['manufacturer'].mask(tiki_work['manufacturer'].isna() == True, 'Unknown')
tiki_work['manufacturer'].isna().sum()
```

```python
tiki_work['manufacturer']
```
